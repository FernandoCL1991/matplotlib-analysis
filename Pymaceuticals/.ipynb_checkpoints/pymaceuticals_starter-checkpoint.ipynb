## Observations and Insights 

```python
# Dependencies and Setup

import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import seaborn as sns
from scipy.stats import linregress
from scipy import stats

# Study data files

mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results

mouse_metadata_df = pd.read_csv(mouse_metadata_path)
study_results_df = pd.read_csv(study_results_path)

# Combine the data into a single dataset

combined_results_df=pd.merge(study_results_df,mouse_metadata_df, how="left",on="Mouse ID")

# Display the data table for preview

combined_results_df
```

```python
# Checking the number of mice.

total_unique_mice=len(combined_results_df["Mouse ID"].unique())
print(f"There were a total of {total_unique_mice} mouses used for the study.")
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 

duplicate_ids = combined_results_df.loc[combined_results_df.duplicated(subset=["Mouse ID", "Timepoint"]), "Mouse ID"].unique()
duplicate_ids
```

```python
# Optional: Get all the data for the duplicate mouse ID.
# 1st Option to get result:

g989_summary=combined_results_df[combined_results_df["Mouse ID"]=="g989"]
g989_summary
```

```python
# Optional: Get all the data for the duplicate mouse ID.
# 2nd Option to get result:

g989_data=combined_results_df.loc[combined_results_df['Mouse ID']== "g989",['Timepoint', 'Tumor Volume (mm3)', 'Metastatic Sites','Drug Regimen', 'Sex', 'Age_months', 'Weight (g)']]
g989_data
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.

cleaned_results_df=combined_results_df.loc[combined_results_df['Mouse ID']!='g989']
cleaned_results_df
```

```python
# Checking the number of mice in the clean DataFrame.

real_number_mice=len(cleaned_results_df["Mouse ID"].unique())
print(f"There were a real total of {real_number_mice} mouses used for the experiment.")
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# 1. Creating Columns and Variables

drug_mean=cleaned_results_df.groupby(['Drug Regimen']).mean()['Tumor Volume (mm3)'].rename('Tumor Volume: Mean')
drug_median=cleaned_results_df.groupby(['Drug Regimen']).median()['Tumor Volume (mm3)'].rename('Tumor Volume: Median')
drug_variance=cleaned_results_df.groupby(['Drug Regimen']).var()['Tumor Volume (mm3)'].rename('Tumor Volume: Variance')
drug_stddev=cleaned_results_df.groupby(['Drug Regimen']).std()['Tumor Volume (mm3)'].rename('Tumor Volume: Standard Deviation')
drug_sem=cleaned_results_df.groupby(['Drug Regimen']).sem()['Tumor Volume (mm3)'].rename('Tumor Volume: Std. Error of Mean')

# 2. Creating New DF based on Variables

drug_df=pd.DataFrame({'Tumor Volume: Mean':drug_mean,\
                     'Tumor Volume: Median':drug_median,\
                     'Tumor Volume: Variance':drug_variance,\
                     'Tumor Volume: Standard Deviation':drug_stddev,\
                     'Tumor Volume: Std. Error of Mean':drug_sem})

# 3. Formatting Values
drug_df['Tumor Volume: Mean']=drug_df['Tumor Volume: Mean'].map("{:,.2f}".format)
drug_df['Tumor Volume: Median']=drug_df['Tumor Volume: Median'].map("{:,.2f}".format)
drug_df['Tumor Volume: Variance']=drug_df['Tumor Volume: Variance'].map("{:,.2f}".format)
drug_df['Tumor Volume: Standard Deviation']=drug_df['Tumor Volume: Standard Deviation'].map("{:,.2f}".format)
drug_df['Tumor Volume: Std. Error of Mean']=drug_df['Tumor Volume: Std. Error of Mean'].map("{:,.2f}".format)

drug_df
```

```python
# Using the aggregation method, produce the same summary statistics in a single line

summary_table_drug = cleaned_results_df.groupby("Drug Regimen").agg({"Tumor Volume (mm3)"\
                                                                     :["mean", "median", "var",\
                                                                       "std", "sem"]})
summary_table_drug
```

# Quartiles, Outliers and Boxplots

```python
# Put treatments into a list for for loop (and later for plot labels)
treatment_list = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

# Create empty list to fill with tumor vol data (for plotting)
all_tumor_vol_data=[]

# Calculate the IQR and quantitatively determine if there are any potential outliers using upper and lower bounds
# Locate the rows which contain mice on each drug and get the tumor volumes
# add subset 
# Determine outliers using upper and lower bounds
    
for x in treatment_list:
    volume=cleaned_results_df.loc[cleaned_results_df['Drug Regimen']==x,'Tumor Volume (mm3)']
    all_tumor_vol_data.append(volume)
    quantile=volume.quantile([.25,.5,.75])
    lower=quantile[.25]
    upper=quantile[.75]
    iqr=upper-lower
    lower_bound = lower - (1.5*iqr)
    upper_bound = upper + (1.5*iqr)
    outliers = volume.loc[(volume< lower_bound) | (volume > upper_bound)]
    print(f'{x} : {outliers.tolist()}')
    print(f'There were a total of {len(outliers.tolist())} of outliers for the {x} treatment')
  
```

```python
# Start by getting the last (greatest) timepoint for each mouse
max_tumor = cleaned_results_df.groupby(["Mouse ID"])['Timepoint'].max()
max_tumor = max_tumor.reset_index()

# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
merged_data = max_tumor.merge(cleaned_results_df,on=['Mouse ID','Timepoint'],how="left")
merged_data
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest

volumes = (all_tumor_vol_data)
fig1, ax1 = plt.subplots()
ax1.boxplot(volumes)
labels=['Capomulin','Ramicane','Infubinol','Ceftamin']
ax1.set_xticklabels(labels)
ax1.set_title('Final Tumor Vol. for Each Mouse Across For Regimens: Capomulin, Ramicane, Infubinol, Ceftamin')
ax1.set_ylabel('Tumor Volume (mm3)')
ax1.set_xlabel("Regimen Type")
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a mouse treated with Capomulin

capomulin_table = cleaned_results_df.loc[cleaned_results_df['Drug Regimen'] == "Capomulin"]
mousedata = capomulin_table.loc[capomulin_table['Mouse ID']== 'r944']
plt.plot(mousedata['Timepoint'],mousedata['Tumor Volume (mm3)'])
plt.xlabel('Timepoint (days)')
plt.ylabel('Tumor Volume (mm3)')
plt.title('Capomulin treatment of mouse r944')
plt.show()
```

```python
# Scatter PLot: Capomulin Drug Regimen
# Tumor Volume vs. Mouse Weight

volume_mean = capomulin_table.groupby('Mouse ID')['Tumor Volume (mm3)'].mean()
weight_mean= capomulin_table.groupby('Mouse ID')['Weight (g)'].mean()

plt.scatter(weight_mean,volume_mean)
plt.title('Average Tumor Size vs. Average Weight for Capomulin Drug Regimen')
plt.ylabel('Average Tumor Size')
plt.xlabel('Average Weight')

plt.show()
```

```python
stats.linregress(volume_mean, weight_mean)
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen

volume_mean = capomulin_table.groupby('Mouse ID')['Tumor Volume (mm3)'].mean()
weight_mean= capomulin_table.groupby('Mouse ID')['Weight (g)'].mean()

(slope, intercept, rvalue, pvalue, stderr) = linregress(weight_mean,volume_mean)
regress_values = weight_mean * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.plot(weight_mean,regress_values,"r-")
plt.annotate(line_eq,(10,10),fontsize=20,color="red")

plt.scatter(weight_mean,volume_mean)
plt.title('Average Tumor Size vs. Weight for Capomulin')
plt.ylabel('Average Tumor Size')
plt.xlabel('Weight')

plt.show()
```

```python
corr = st.pearsonr(capomulin_table['Weight (g)'],capomulin_table['Tumor Volume (mm3)'])
print(corr[0])
```

```python
print(f'There is {corr[0]} correlation between the Average Tumor Size and Weight when using the Capomulin Drug Regimen.')
```
